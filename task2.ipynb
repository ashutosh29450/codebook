```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data2.json")
```

```python
data
```

```python
def clean_data(data):
    clean_user_data = []
    for user in data['users']:
        if(user['name'] != ""):
            s = set()
            for friend in user['friends']:
                s.add(friend)
            cleaned_friends = []
            for x in s:
                cleaned_friends.append(x)
            user["friends"] = cleaned_friends
            clean_user_data.append(user)
            
    data['users'] = clean_user_data

    clean_page_data = []
    s = set()
    for page in data['pages']:
        if page['id'] in s:
            continue
        else:
            s.add(page['id'])
            clean_page_data.append(page)
    data['pages'] = clean_page_data
    return data
```

```python
data = clean_data(data)
data
```
