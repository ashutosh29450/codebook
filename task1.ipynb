```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
def display_users(data):
    for user in data["users"]:
        print(f"{user['name']}")
```

```python
display_users(data)
```

```python
def display_users_and_friends(data):
    for user in data["users"]:
        s = ""
        s = s + user["name"]
        s += " is friends with : "
        for friend in user["friends"]:
            s += str(friend)
            s += " "
        print(s)
```

```python
display_users_and_friends(data)
```

```python
def display_pages(data):
    for page in data["pages"]:
        s = ""
        s += str(page["id"])
        s += " : "
        s += page["name"]
        print(s)
```

```python
display_pages(data)
```
