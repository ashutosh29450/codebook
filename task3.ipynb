```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data3.json")
data
```

```python
id_to_name: dict[int, str] = {}
for user in data['users']:
    id_to_name[user['id']] = user['name']
id_to_name
```

```python
from collections import defaultdict

id_to_friend_ids: dict[int, set[int]] = defaultdict(set)
for user in data['users']:
    for friend in user['friends']:
        id_to_friend_ids[user['id']].add(friend)
id_to_friend_ids
```

```python
from collections import defaultdict

id_to_mayknowid: dict[int, set[int]] = defaultdict(set)

for key, value in list(id_to_friend_ids.items()):
    for friends in value:
        for known in id_to_friend_ids.get(friends, set()):
            id_to_mayknowid[key].add(known)

id_to_mayknowid       
```

```python
for key, value in id_to_mayknowid.items():
    value.discard(key)
id_to_mayknowid
```
